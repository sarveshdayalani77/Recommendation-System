**Calculating Cosine Similarity between users**

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(df.values,metric='cosine')
```

```python
user_sim
```

```python
df_user=pd.DataFrame(user_sim)
```

```python
df_user.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
df_user.iloc[0:5, 0:5]
```

```python
#Most Similar Users
df_user.idxmax(axis=1)[0:5]
```

```python
#Most Similar Users
print(df_user.idxmax(axis=1)[1348])
print(df_user.max(axis=1).sort_values(ascending=False).head(50))
```
